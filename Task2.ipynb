```python
import numpy as np
import matplotlib.pyplot as plt

# 1.  First derivative
# f(x)=(x**4)/4 -x**2 -x
# Forward difference
# y_n = (y_{n+1} - y_n)/h 
# h>0

# 2. First derivative
# f(x)= (x**4)/4 -x**2 -x
# Backward difference
# y_n = (y_n - y_{n-1})/h 
# h<0

# 3. First derivative
# f(x)= (x**4)/4 -x**2 -x
# Centered difference
# y_n = (y_{n+1} - y_{n-1})/h 
# h=0


x=np.linspace(-2,3,100) 
x1=np.linspace(-5,5,100)

H_step=0.1 

def f(x):
    y=1/4*x**4-x**2-x
    return y

def g(x):
    w = np.tanh(x)
    return w

#Foward difference 
def forward(f,h,x):
    y_p=(f(x+h)-f(x))/ h
    return y_p

#Derivative back difference
def backward(f,h,x):
    y_p=(f(x)-f(x-h))/h
    return y_p

#First centred difference
def centered(f,h,x):
    y_p=(f(x+h)-f(x-h))/2*h
    return y_p

#for equation a)
#Plotting graphs of function a) in the same axes
plt.figure()
plt.plot(x,x**3-2*x-1,'k',label = 'Theoretical')  #Plotting the derivative of our function
plt.plot(x,forward(f,h_step,x),'.r',label = 'Numerical_Forward')  
plt.plot(x,backward(f,h_step,x),'ob',label = 'Numerical_Backward')
plt.plot(x,centered(f,h_step,x),'.y',label = 'Numerical_Centered')
plt.legend()

#error
plt.figure()
plt.plot(x,x**3-2*x-1 - forward(f,h_step,x),'*')
plt.plot(x,x**3-2*x-1 - backward(f,h_step,x),'-')
plt.plot(x,x**3-2*x-1 -centered(f,h_step,x) ,'--')

#for equation b)
#Plotting graphs of function a) in the same axes
plt.figure()
plt.plot(x1,1/(np.exp(x)+np.exp(-x))**2,'k',label = 'Theoretical')  #Plotting the derivative of our function
plt.plot(x1,forward(g,h_step,x1),'.r',label = 'Numerical_Forward')  
plt.plot(x1,backward(g,h_step,x1),'ob',label = 'Numerical_Backward')
plt.plot(x1,centered(g,h_step,x1),'.y',label = 'Numerical_Centered')
plt.legend()

#error
plt.figure()
plt.plot(x1,1/(np.exp(x)+np.exp(-x))**2 - forward(g,h_step,x1),'*')
#plt.plot(x1,1/(np.exp(x)+np.exp(-x))**2 - backward(g,h_step,x1),'-'))
plt.plot(x1,1/(np.exp(x)+np.exp(-x))**2 - backward(g,h_step,x1),'--')
plt.plot(x1,1/(np.exp(x)+np.exp(-x))**2 -centered(g,h_step,x1) ,'--')

#Forward and backward are above the theoretical whichh is not accurate
#The centered forms a straight line which we cant compare to any
```
